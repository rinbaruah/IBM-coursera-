### The code below is a function that finds whether a number is odd or even.

```python
def odd_even(num):
    
    if num%2==0:
        print('{} is an even number'.format(num))
    else:
        if num==1:
            print('{} is neither odd, nor even'.format(num))
        else:
            
            print('{} is an odd number'.format(num))
```

```python
odd_even(4)
```

```python
odd_even(3)
```

```python
odd_even(1)
```
